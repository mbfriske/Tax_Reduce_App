```python
import pandas as pd 
import numpy as np
import os
```

## Import CSV Files 

```python
# Account Info CSV - data for each appraisal account 
account_info = pd.read_csv("../DCAD/account_info.csv")
account_info.head()
```

```python
# Residential Details CSV
res_detail = pd.read_csv("../DCAD/res_detail.csv")
res_detail.head()
```

```python
res_addl = pd.read_csv("../DCAD/res_addl.csv", encoding="cp1252")

res_addl.head()
```

## Merge CSV files into DataFrame

```python
# Merge res_addl & res_detail on ACCOUNT_NUM
merge_one = pd.merge(res_addl, res_detail, on='ACCOUNT_NUM')
merge_one.head()
```

```python
# Clean merge_one Columns 
merge_one.columns
```

```python
# Rename columns so that they are differentiated
merge_one = merge_one.rename(columns={"APPRAISAL_YR_x": "APPRAISAL_YR", "TAX_OBJ_ID_x": "TAX_OBJ_ID", "YR_BUILT_x": "YR_BUILT",
                                      "EXT_WALL_DESC_x": "EXT_WALL_DESC", "DEPRECIATION_PCT_x": "DEPRECIATION_PCT"})

merge_one = merge_one.rename(columns={"APPRAISAL_YR_y": "APPRAISAL_YR", "TAX_OBJ_ID_y": "TAX_OBJ_ID", "YR_BUILT_y": "YR_BUILT",
                                      "EXT_WALL_DESC_y": "EXT_WALL_DESC", "DEPRECIATION_PCT_y": "DEPRECIATION_PCT"})

merge_one.head()
```

```python
# Merge merge_one & account_info in DataFrame 
final_df = pd.merge(merge_one, account_info, on='ACCOUNT_NUM')
final_df.head()
```

```python
# Clean final_df Columns 
final_df.columns
```

```python
# Rename columns so that they are differentiated
final_df = final_df.rename(columns={"APPRAISAL_YR_x": "APPRAISAL_YR"})

final_df = final_df.rename(columns={"APPRAISAL_YR_y": "APPRAISAL_YR"})

final_df.head()
```

```python
# Check columns 
final_df.columns
```

```python
final_df.describe()
```

```python
# Export to CSV
final_df.to_csv('Appraisal_Data.csv', index=False)
```
